# Import packages :

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time 
import matplotlib.pyplot as plt
```

# Scrape the website and retrieve the data :

```python
BA_url = 'https://www.airlinequality.com/airline-reviews/british-airways'


Nb_pages = 36
page_size = 100

Comment = []

# for i in range(1, pages + 1):
for i in range(1, Nb_pages + 1):
    time.sleep(2)
    print(f"Scraping page {i}")

    # Create URL to collect links from paginated data
    url = f"{BA_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        Comment.append(para.get_text())
    
    print(f"   ---> {len(Comment)} total reviews")

```

```python
# Load the data in a csv file 

df = pd.DataFrame()
df["reviews"] = Comment
df.to_csv('Reviews_British_airways.csv')

# Print the head of the dataframe
df.head()
```

# Data processing :

```python
#importation des données à partir du fichier csv

df = pd.read_csv('Reviews_British_airways.csv')
df.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')
```

```python
# Nettoyer les données 
df['reviews'] = df['reviews'].str.replace('✅ Trip Verified |','')
df['reviews'] = df['reviews'].str.replace('|','')
df['reviews'] = df['reviews'].str.replace('\r\n','')
df
```

```python
# permet la manipukation des chaines de caractéres 
import re 

# (Natural Language Toolkit) pour le traitement du loangage naturel 
import nltk

# Les stopwords sont des mots couramment utilisés dans une langue  
#(comme "le", "et", "mais", etc.) qui sont généralement considérés comme non informatifs lors de l'analyse de texte
from nltk.corpus import stopwords

# Le lemme est une forme canonique ou une base commune d'un mot. Le WordNetLemmatizer est utilisé pour lemmatiser des mots
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt') # est utilisé pour diviser un texte en phrases et en mots individuels
nltk.download('stopwords')
nltk.download('wordnet')# WordNet fournit des synonymes, des antonymes et des relations lexicales entre les mots

# Conversion en minuscules
df['reviews1'] = df['reviews'].str.lower()

# Suppression de la ponctuation et des caractères spéciaux 
df['reviews1'] = df['reviews1'].apply(lambda x: re.sub(r'[^\w\s]', '', x))

# Tokenisation
df['reviews1'] = df['reviews1'].apply(nltk.word_tokenize)

# Suppression des mots vides
stop_words = set(stopwords.words('english'))
df['reviews1'] = df['reviews1'].apply(lambda x: [word for word in x if word not in stop_words])

# Lemmatisation
lemmatizer = WordNetLemmatizer()
df['reviews1'] = df['reviews1'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

df
```

* Conversion en minuscules : Le texte est converti en minuscules pour normaliser la casse des mots.

* Suppression de la ponctuation et des caractères spéciaux : Tous les signes de ponctuation et les caractères spéciaux sont supprimés du texte. Cela permet de se concentrer uniquement sur les mots et d'éliminer les éléments non alphabétiques.

* Tokenisation : Le texte est divisé en mots individuels ou en "tokens". Les phrases sont séparées en fonction des espaces blancs et des signes de ponctuation, et chaque mot est considéré comme un élément distinct.

* Suppression des stopwords : Les stopwords, tels que "le", "et", "mais", etc., qui sont des mots couramment utilisés et peu informatifs, sont supprimés du texte. Cela permet de réduire le bruit et de se concentrer sur les mots clés plus significatifs.

* Lemmatisation : Les mots restants sont réduits à leur forme canonique ou base commune en utilisant la lemmatisation. Par exemple, les mots "courir", "courais" et "couru" peuvent être réduits à "courir". Cela permet de normaliser les mots et de regrouper les formes similaires.



```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Calcul du score de sentiment pour chaque commentaire
df['sentiment_score'] = df['reviews1'].apply(lambda x: sia.polarity_scores(' '.join(x))['compound'])
```

```python
df
```

```python
# count des messages positifs ou négatifs et les neutres
# le seuille est fixé entre -0.25 et 0.25 

# Fonction pour déterminer la catégorie du sentiment
def get_sentiment_category(score):
    if score >= -0.25 and score <= 0.25:
        return 'Neutral'
    elif score > 0.25:
        return 'Positive'
    else:
        return 'Negative'

# Création de la nouvelle colonne 'sentiment_category' en appliquant la fonction à chaque valeur de sentiment_score
df['sentiment_category'] = df['sentiment_score'].apply(get_sentiment_category)
df
```

```python
# Calculer le nombre de occurrences de chaque catégorie de sentiment
sentiment_counts = df['sentiment_category'].value_counts()
print(sentiment_counts)
# Calculer les pourcentages en divisant par le nombre total d'observations
sentiment_percentages = (sentiment_counts / len(df)) * 100
print(sentiment_percentages)
```

```python
# Create a pie chart
plt.pie(sentiment_percentages, labels=sentiment_percentages.index, autopct='%1.1f%%')

# Add a title
plt.title('Percentage of each sentiment category')

# Display the chart
plt.show()
```

```python
# Afficher le topic de chaque catégorie de sentiment 
from gensim import corpora, models


positive_df = df.loc[df['sentiment_category'] == 'Positive']
negative_df = df.loc[df['sentiment_category'] == 'Negative']
neutre_df = df.loc[df['sentiment_category'] == 'Neutral']
# Création du dictionnaire à partir des commentaires
dictionary = corpora.Dictionary(positive_df['reviews1'])

# Transformation des commentaires en sacs de mots (bow = Bag of Words)
corpus = [dictionary.doc2bow(comment) for comment in positive_df['reviews1']]

# Modélisation des sujets avec LDA
lda_model = models.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=10)

# Affichage des sujets et leurs mots clés
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_wordcloud(df,title):
    # Concaténation de tous les commentaires en une seule chaîne de texte
    all_text = ' '.join([' '.join(comment) for comment in df['reviews1']])
    
    # Création du nuage de mots
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
    
    # Affichage du nuage de mots
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()

generate_wordcloud(positive_df,'Positive topics')
generate_wordcloud(negative_df,'Negative topics')
generate_wordcloud(neutre_df,'Neutre topics')
```
